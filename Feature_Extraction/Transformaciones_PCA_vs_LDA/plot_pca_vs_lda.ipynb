# Comparison of LDA and PCA 2D projections

```python
#NumPy: numerical scientific computing 
import numpy as np

#Scikit-learn: machine learning
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

#Matplotlib: plotting 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
#%matplotlib notebook

#Pandas: python data analysis
import pandas #as pd
#Seaborn: statistical data visualization
import seaborn #as sns
seaborn.set(color_codes=True)

#SciPy: scientific tools
import scipy.io.matlab as matlab
```

```python
# Lectura de fichero de Matlab .mat
mat_file =  "extrac_train.mat"
mat = matlab.loadmat(mat_file,squeeze_me=True) #devuelve un dictionary
#print(list(mat.keys())) #variables almacenadas

# Lectura de los datos
X = mat["x1"]   
y = mat["clab1"]   
del mat
print(X.shape,y.shape)
```

```python
#Transformaciones PCA y LDA
pca = PCA(n_components = 2, svd_solver='full')
pca.fit(X)
Xpca = pca.transform(X)

lda = LDA()
lda.fit(X,y)
Xlda = lda.transform(X)
```

```python
#Representacion datos transformados
plt.figure()
plt.subplot(121)
plt.scatter(Xpca[:, 0], Xpca[:, 1], c=y,cmap='jet')
plt.title('PCA')
plt.subplot(122)
plt.scatter(Xlda[:, 0], Xlda[:, 1], c=y,cmap='jet')
plt.title('LDA')
```

Aunque LDA es un algoritmo supervisado los resultados no son muy impresionantes, de hecho no está claro que haya funcionado mejor que simplemente utilizar una PCA. En este conjunto de datos y según el tipo de clasificación que se busque, puede ser más interesante utilizar PCA como herramienta de reducción de dimensionalidad.
