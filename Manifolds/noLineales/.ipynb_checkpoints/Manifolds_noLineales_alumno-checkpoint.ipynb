## 1. Embedding con los 4 métodos.

```python
from sklearn.manifold import LocallyLinearEmbedding, SpectralEmbedding,MDS, Isomap
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import manhattan_distances, euclidean_distances
from mpl_toolkits.mplot3d import Axes3D
from sklearn.utils import check_random_state
from sklearn import manifold, datasets
import numpy as np
Axes3D

# ----------------------------------------------------------------------
# Swiss roll   n_samples=1500

X1, color = datasets.make_swiss_roll(n_samples=500)

# ----------------------------------------------------------------------
# Sphere      n_samples=1000

n_samples = 300

# Create our sphere.
random_state = check_random_state(0)
p = random_state.rand(n_samples) * (2 * np.pi - 0.55)
t = random_state.rand(n_samples) * np.pi

# Sever the poles from the sphere.
indices = (t < (np.pi - (np.pi / 8))) & (t > ((np.pi / 8)))
colors = p[indices]
x, y, z = (
    np.sin(t[indices]) * np.cos(p[indices]),
    np.sin(t[indices]) * np.sin(p[indices]),
    np.cos(t[indices]),
)

X2 = np.array([x, y, z]).T


# ----------------------------------------------------------------------
# Locally linear embedding of the swiss roll

print("Computing Spectral embedding ()")

model1 = SpectralEmbedding(n_components=2,n_neighbors=12)
X1_t_SE=model1.fit_transform(X1)


model2 = SpectralEmbedding(n_components=2,n_neighbors=12)
X2_t_SE=model2.fit_transform(X2)


# ----------------------------------------------------------------------
# Locally linear embedding of the swiss roll

print("Computing LLE embedding")

model1 = LocallyLinearEmbedding(n_components=2,n_neighbors=12)
X1_t_LLE=model1.fit_transform(X1)


model2 = LocallyLinearEmbedding(n_components=2,n_neighbors=12)
X2_t_LLE=model2.fit_transform(X2)
plt.scatter(X2_t_LLE[:,0],X2_t_LLE[:,1],c=colors)
plt.show()

# ----------------------------------------------------------------------
# Isomap of the swiss roll

print("Computing Isomap embedding")

model1 = Isomap(n_components=2,n_neighbors=12)
X1_t_ISO=model1.fit_transform(X1)


model2 = Isomap(n_components=2,n_neighbors=12)
X2_t_ISO=model2.fit_transform(X2)
plt.scatter(X2_t_ISO[:,0],X2_t_ISO[:,1],c=colors)
plt.show()

# ----------------------------------------------------------------------
# MDS of the swiss roll

print("Computing MDS embedding")

model1 = MDS(n_components=2)
X1_t_MDS=model1.fit_transform(X1)


model2 = MDS(n_components=2)
X2_t_MDS=model2.fit_transform(X2)
plt.scatter(X2_t_MDS[:,0],X2_t_MDS[:,1],c=colors)
plt.show()


# -----------------------------------------------------------------------

# Plot results

fig = plt.figure(figsize=(15, 13))

ax1 = fig.add_subplot(251, projection="3d")
ax1.scatter(X1[:, 0], X1[:, 1], X1[:, 2], c=color, cmap=plt.cm.Spectral)
ax1.set_title("Rollo suizo")

ax2 = fig.add_subplot(252)
ax2.scatter(X1_t_SE[:,0],X1_t_SE[:,1],c=color,cmap=plt.cm.Spectral)
ax2.set_title("Gráfico 2")

ax3 = fig.add_subplot(253)
ax3.scatter(X1_t_LLE[:,0],X1_t_LLE[:,1],c=color,cmap=plt.cm.Spectral)
ax3.set_title("Gráfico 3")

ax4 = fig.add_subplot(254)
ax4.scatter(X1_t_ISO[:,0],X1_t_ISO[:,1],c=color,cmap=plt.cm.Spectral)
ax4.set_title("Gráfico 4")

ax5 = fig.add_subplot(255)
ax5.scatter(X1_t_MDS[:,0],X1_t_MDS[:,1],c=color,cmap=plt.cm.Spectral)
ax5.set_title("Gráfico 5")

plt.show()







fig = plt.figure(figsize=(15, 13))


ax = fig.add_subplot(151, projection="3d")
ax.scatter(x, y, z, c=p[indices], cmap=plt.cm.rainbow)
ax.view_init(40, -10)
ax.set_title("Esfera")


ax2 = fig.add_subplot(252)
ax2.scatter(X2_t_SE[:,0],X2_t_SE[:,1],c=colors,cmap=plt.cm.rainbow)
ax2.set_title("Gráfico 2")

ax3 = fig.add_subplot(253)
ax3.scatter(X1_t_LLE[:,0],X1_t_LLE[:,1],c=color,cmap=plt.cm.Spectral)
ax3.set_title("Gráfico 3")

ax4 = fig.add_subplot(254)
ax4.scatter(X1_t_ISO[:,0],X1_t_ISO[:,1],c=color,cmap=plt.cm.Spectral)
ax4.set_title("Gráfico 4")

ax5 = fig.add_subplot(255)
ax5.scatter(X1_t_MDS[:,0],X1_t_MDS[:,1],c=color,cmap=plt.cm.Spectral)
ax5.set_title("Gráfico 5")

plt.show()



```
