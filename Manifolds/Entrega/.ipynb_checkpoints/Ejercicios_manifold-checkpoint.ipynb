#### t-SNE con y3 en 2D

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

early_ex = 20


tsne_per5 = TSNE(n_components=2,perplexity=5,early_exaggeration=early_ex,random_state=0)
tsne_per10 = TSNE(n_components=2,perplexity=10,early_exaggeration=early_ex,random_state=0)
tsne_per15 = TSNE(n_components=2,perplexity=15,early_exaggeration=early_ex,random_state=0)
tsne_per20 = TSNE(n_components=2,perplexity=20,early_exaggeration=early_ex,random_state=0)
tsne_per25 = TSNE(n_components=2,perplexity=25,early_exaggeration=early_ex,random_state=0)
tsne_per30 = TSNE(n_components=2,perplexity=30,early_exaggeration=early_ex,random_state=0)

# Aplicar t-SNE al conjunto de datos
X_tsne_per5 = tsne_per5.fit_transform(X)
X_tsne_per10 = tsne_per10.fit_transform(X)
X_tsne_per15 = tsne_per15.fit_transform(X)
X_tsne_per20 = tsne_per20.fit_transform(X)
X_tsne_per25 = tsne_per25.fit_transform(X)
X_tsne_per30 = tsne_per30.fit_transform(X)

# Plotear los datos reducidos por t-SNE



fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(10, 7))

axs[0,0].scatter(X_tsne_per5[:, 0], X_tsne_per5[:, 1],c=y3)
axs[0,0].set_title('perplexity=5')

axs[0,1].scatter(X_tsne_per10[:, 0], X_tsne_per10[:, 1],c=y3)
axs[0,1].set_title('perplexity=10')

axs[0,2].scatter(X_tsne_per15[:, 0], X_tsne_per15[:, 1],c=y3)
axs[0,2].set_title('perplexity=15')

axs[1,0].scatter(X_tsne_per20[:, 0], X_tsne_per20[:, 1],c=y3)
axs[1,0].set_title('perplexity=20')

axs[1,1].scatter(X_tsne_per25[:, 0], X_tsne_per25[:, 1],c=y3)
axs[1,1].set_title('perplexity=25')

axs[1,2].scatter(X_tsne_per30[:, 0], X_tsne_per30[:, 1],c=y3)
axs[1,2].set_title('perplexity=30')


plt.show()
```

Obtenemos resultados bastante parecidos para todos los valores de perplexity, vamos a optar por utilizar una perplexity de 20.
Los conjuntos están bastante separados y los puntos de una clase que se mezclan con los de otra aparecen con cualquier nivel de perplexity.

#### UMAP con y3  en 2D

```python
from umap import UMAP


umap_nei5 = UMAP(densmap= True,n_components=2,n_neighbors=5,random_state=0)
umap_nei10 = UMAP(densmap= True,n_components=2,n_neighbors=10,random_state=0)
umap_nei15 = UMAP(densmap= True,n_components=2,n_neighbors=15,random_state=0)
umap_nei20 = UMAP(densmap= True,n_components=2,n_neighbors=20,random_state=0)
umap_nei25 = UMAP(densmap= True,n_components=2,n_neighbors=25,random_state=0)
umap_nei30 = UMAP(densmap= True,n_components=2,n_neighbors=30,random_state=0)

# Aplicar umap al conjunto de datos
X_umap_nei5 = umap_nei5.fit_transform(X)
X_umap_nei10 = umap_nei10.fit_transform(X)
X_umap_nei15 = umap_nei15.fit_transform(X)
X_umap_nei20 = umap_nei20.fit_transform(X)
X_umap_nei25 = umap_nei25.fit_transform(X)
X_umap_nei30 = umap_nei30.fit_transform(X)

# Plotear los datos reducidos por umap

fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(10, 7))

axs[0,0].scatter(X_umap_nei5[:, 0], X_umap_nei5[:, 1],c=y3)
axs[0,0].set_title('n_neigbours=5')

axs[0,1].scatter(X_umap_nei10[:, 0], X_umap_nei10[:, 1],c=y3)
axs[0,1].set_title('n_neigbours=10')

axs[0,2].scatter(X_umap_nei15[:, 0], X_umap_nei15[:, 1],c=y3)
axs[0,2].set_title('n_neigbours=15')

axs[1,0].scatter(X_umap_nei20[:, 0], X_umap_nei20[:, 1],c=y3)
axs[1,0].set_title('n_neigbours=20')

axs[1,1].scatter(X_umap_nei25[:, 0], X_umap_nei25[:, 1],c=y3)
axs[1,1].set_title('n_neigbours=25')

axs[1,2].scatter(X_umap_nei30[:, 0], X_umap_nei30[:, 1],c=y3)
axs[1,2].set_title('n_neigbours=30')


plt.show()
```

Vistos los resultados, para la clasificación de las clases _y3_ vamos a tomar _n\_neigbours_ = 10. De esta forma no requerimos de demasiado tiempo de computación y las tres clases están (dentro de lo posible) bastante separadas.

```python
from time import time

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import offsetbox
from sklearn import (manifold, datasets, decomposition, ensemble,
                     discriminant_analysis, random_projection)
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.manifold import LocallyLinearEmbedding, SpectralEmbedding,MDS, Isomap
```

#### 3.1 Estimación de los parámetros

```python
print("Computing Spectral embedding ")

st = time()

SpecEmb = SpectralEmbedding(n_components=3,n_neighbors=30)
X_SpecEmb = SpecEmb.fit_transform(X)

et = time()

elapsed_time = et - st
print('Execution time:', elapsed_time, 'seconds')

fig = plt.figure(figsize=(15,6))
plt.suptitle('Spectral Embedding con n_neighbors=30', fontsize=16)

# Hacer el scatter plot en 3D
ax = fig.add_subplot(121, projection='3d')
ax.scatter(X_SpecEmb[:, 0], X_SpecEmb[:, 1], X_SpecEmb[:, 2],c=y3)
ax.set_title('Separados por región')

ax = fig.add_subplot(122, projection='3d')
ax.scatter(X_SpecEmb[:, 0], X_SpecEmb[:, 1], X_SpecEmb[:, 2],c=y9)
ax.set_title('Separados por área')


plt.show()
```

```python
print("Computing Locally Linear Embedding ")

st = time()

LLE = LocallyLinearEmbedding(n_components=3,n_neighbors=30)
X_LLE = LLE.fit_transform(X)

et = time()

elapsed_time = et - st
print('Execution time:', elapsed_time, 'seconds')

fig = plt.figure(figsize=(15,6))
plt.suptitle('Locally Linear Embedding con n_neighbors=30', fontsize=16)

# Hacer el scatter plot en 3D
ax = fig.add_subplot(121, projection='3d')
ax.scatter(X_LLE[:, 0], X_LLE[:, 1], X_LLE[:, 2],c=y3)
ax.set_title('Separados por región')

ax = fig.add_subplot(122, projection='3d')
ax.scatter(X_LLE[:, 0], X_LLE[:, 1], X_LLE[:, 2],c=y9)
ax.set_title('Separados por área')


plt.show()
```
